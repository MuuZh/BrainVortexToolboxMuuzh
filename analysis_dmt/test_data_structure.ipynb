```python
import pandas as pd
import pyarrow
# 定义处理单个spiral数据的函数
def process_spiral(spiral_data):
    df = pd.DataFrame({
        'sub_id': spiral_data['sub_id'],
        'frame': spiral_data['frame'],
        'center_coordinate': spiral_data['center_coordinate'],
        'span_coordinates': spiral_data['span_coordinates']
    })
    df['spiral_id'] = spiral_data['spiral_id']  # 添加spiral_id列
    return df

# 假设spiral_list是一个包含所有spirals数据的列表
spiral_list = [
    # 示例数据，实际数据应按此格式填充
    {   
        'sub_id': 1, 
        'spiral_id': 100,
        'frame': [81, 82, 83, 84, 85, 86],
        'center_coordinate': [(20,30), (21,31), (22,31), (25,32), (24,32), (28,35)],
        'span_coordinates': [
            [(20, 30), (19, 30), (20, 31)],
            [(21, 31), (20, 31), (21, 32)],
            [(22, 31), (21, 32), (22, 33)],
            [(25, 32), (24, 32), (25, 33)],
            [(24, 32), (23, 32), (24, 33)],
            [(28, 35), (27, 35), (28, 36)]
        ]
    },
    {   
        'sub_id': 2,
        'spiral_id': 101,
        'frame': [81, 82, 83, 84, 85, 86],
        'center_coordinate': [(20,30), (21,31), (22,31), (25,32), (24,32), (28,35)],
        'span_coordinates': [
            [(20, 30), (19, 30), (20, 31)],
            [(21, 31), (20, 31), (21, 32)],
            [(22, 31), (21, 32), (22, 33)],
            [(25, 32), (24, 32), (25, 33)],
            [(24, 32), (23, 32), (24, 33)],
            [(28, 35), (27, 35), (28, 36)]
        ]
    }
    # 更多spiral数据
]

# 初始化空DataFrame
all_data = pd.DataFrame()

# 迭代spiral_list，处理每一个spiral的数据，并追加到all_data
for spiral_data in spiral_list:
    df = process_spiral(spiral_data)
    all_data = pd.concat([all_data, df], ignore_index=True)

# 存储为Parquet文件
all_data.to_parquet('all_spirals.parquet', engine='pyarrow')
```
